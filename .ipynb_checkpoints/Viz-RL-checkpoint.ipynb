```python
import numpy as np
import pandas as pd
import csv
```

```python
with open("DQN_metrics.txt",'r') as f:
    lines = f.readlines()
    print(lines[0])
```

```python
new_lines = []
for line in lines:
    new_lines.append(line[14:])
```

```python
new_dict = []
for line in new_lines:
    dict_row = {}
    new_line = line.split(",")
    dict_row[new_line[4].split(":")[0]] = new_line[4].split(":")[1]; 
    print(dict_row)
    
```

```python
with open("dqn_metrics.csv", 'w') as f:
    fieldnames = ['episode', 'duration', 'episode steps', 'steps per second', 'episode reward', 'mean reward', 'mean action', 'mean observation', 'loss', 'mean_absolute_error', 'mean_q']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    
    writer.writeheader()
    for row in new_lines:
        writer.writerow({row})
```
