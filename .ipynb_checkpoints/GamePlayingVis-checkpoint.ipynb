# Modifying all dataframes to hold same number of rows

```python
df3.tail(5)
```

```python
df2m = df2[:198]
df3m = df3[:198]
```

```python
ax = sns.lineplot(x= 'episode', y = ' loss',data = df)
```

```python
ax1 = df.plot(x='episode',y=' loss')
ax2 = df2m.plot(x='episode',y=' loss',ax=ax1)
ax3 = df3m.plot(x='episode',y=' loss',ax=ax2)
ax2.legend(["DQN", "Duel_DQN","SARSA"])
ax2.set(xlabel="Episode", ylabel="Loss")
```

```python
rew = df.plot(x='episode',y=' episode reward')
rew2 = df2m.plot(x='episode',y=' episode reward',ax=rew)
rew3 = df3m.plot(x='episode',y=' episode reward',ax=rew2)
rew3.set(xlabel="Episode", ylabel="Reward")
rew3.legend(["DQN", "Duel_DQN","SARSA"])
```

```python
loss = df.plot(x='episode',y=' episode steps')
loss2 = df2m.plot(x='episode',y=' episode steps',ax=loss)
loss3 = df3m.plot(x='episode',y=' episode steps',ax=loss2)
loss3.set(xlabel="Episode", ylabel="Steps")
loss3.legend(["DQN", "Duel_DQN","SARSA"])
```

```python
error = df.plot(x='episode',y=' mean_absolute_error')
error2 = df2m.plot(x='episode',y=' mean_absolute_error',ax = error)
error3 = df3m.plot(x='episode',y=' mean_absolute_error',ax = error2)
error3.legend(["DQN", "Duel_DQN","SARSA"])
error3.set(xlabel="Episode", ylabel="Mean error")
```

```python
qval = df.plot(x='episode',y=' mean_q')
qval2 = df2m.plot(x='episode',y=' mean_q',ax = qval)
qval3 = df3m.plot(x='episode',y=' mean_q',ax = qval2)
qval3.legend(["DQN", "Duel_DQN","SARSA"])
qval3.set(xlabel="Episode", ylabel="Mean q")
```
