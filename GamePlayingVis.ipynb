```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_dqn = pd.read_csv("./data/DQN_metrics.csv")
df_ddqn = pd.read_csv("./data/Duel_DQN_metrics.csv")
df_sarsa = pd.read_csv("./data/SARSA_metrics.csv")
```

```python
print(df_dqn.info())
df_dqn.head()
```

```python
print(df_ddqn.info())
df_ddqn.head()
```

```python
print(df_sarsa.info())
df_sarsa.head()
```

```python
fig = plt.figure(num=None, figsize=(16, 6), dpi=80)
plt.grid(True, linewidth=0.5, linestyle='--')
sns.lineplot(x="episode", y=" episode reward", data=df_dqn)
sns.lineplot(x="episode", y=" episode reward", data=df_ddqn)
sns.lineplot(x="episode", y=" episode reward", data=df_sarsa)
plt.legend(["DQN", "Duel_DQN","SARSA"])
plt.axvline(x=232, linestyle="dashed", color = 'black')
fig.suptitle("Reward per episode across all methods", fontsize=12)
plt.xlabel("Episode")
plt.ylabel("Reward")
fig.savefig("Reward per episode.jpg")
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(10, 5), dpi=80)
sns.jointplot(x="episode", y=" loss", data=df_dqn, kind = "kde")
fig.savefig("loss per episode_dqn.jpg")
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(10, 5), dpi=80)
sns.jointplot(x="episode", y=" loss", data=df_ddqn, kind = "kde", color = "darkblue")
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(10, 5), dpi=80)
sns.jointplot(x="episode", y=" loss", data=df_sarsa,kind = "scatter", color = "darkgreen")
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(16, 6), dpi=80)
plt.grid(True, linewidth=0.5, linestyle='--')
sns.lineplot(x="episode", y=" mean_absolute_error", data=df_dqn, color = "red")
sns.lineplot(x="episode", y=" mean_absolute_error", data=df_ddqn, color = "black")
sns.lineplot(x="episode", y=" mean_absolute_error", data=df_sarsa, color = "blue")
plt.legend(["DQN", "Duel_DQN","SARSA"])
plt.axvline(x=232, linestyle="dashed", color = 'black')
fig.suptitle("Absolute error per episode across all methods", fontsize=12)
```

```python
fig = plt.figure(num=None, figsize=(16, 6), dpi=80)
plt.grid(True, linewidth=0.5, linestyle='--')
sns.lineplot(x="episode", y=" mean_q", data=df_dqn, color = "yellow")
sns.lineplot(x="episode", y=" mean_q", data=df_ddqn, color = "turquoise")
sns.lineplot(x="episode", y=" mean_q", data=df_sarsa, color = "darkgreen")
plt.legend(["DQN", "Duel_DQN","SARSA"])
plt.axvline(x=232, linestyle="dashed", color = 'black')
fig.suptitle("Mean_q per episode across all methods", fontsize=12)
```
